# Bivariate Analysis

```python
#correlation matrix
df.corr()
```

```python
fig, axes = plt.subplots(1,1, figsize = (8,5))
sns.heatmap(df.corr())
plt.show()
```

## Here we can see discount and profit are highly negatively corelated

* Values closer to zero means there is no linear trend between the two variables.

* The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increase

```python
sns.catplot(x='Category',y='Profit',data=df,col="Region")
```

```python
plt.figure(figsize=(10,10))
df.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

## From the above graph we can see that the sales of furniture is good still profit is lowest

```python
plt.figure(figsize=(24,10))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize=(24,10))
df.groupby('Sub-Category')['Profit'].agg(['sum']).plot.bar()
plt.show()
```

## From the above figure we can observe that:

* Sale of tables is high but it is producing loss
* There is a negligible loss from Bookcases and suppliers
* There is almost negligible sale and profit from fasteners
* Copiers are giving the highest profit

```python
sns.barplot(x=df.Discount,y=df.Profit)
```

## From the above graph we can see that from more than 20% discount loss is happening

```python
sns.barplot(x=df.Region, y=df.Sales)
```

```python
sns.barplot(x=df.Region, y=df.Profit)
```

```python
sns.barplot(x=df.Region, y=df.Profit,hue=df.Category)
```

## 1) For the central region the Furniture category is making most loss.

## 2) Technology is giving the highest profit in every region

```python
sns.barplot(x=df.Region, y=df.Profit,hue=df.Segment)
```

### From the above figure we can see that the West region produces highest net profit and East region is producing very high profit in Home office segment while the lowest profit was in Central region in consumer segment.

```python
dataplot = df.groupby(['State'])['Sales', 'Profit'].sum()
dataplot.plot.bar(figsize = (20,10))
```

```python
dataplot = df.groupby(['State'])['Profit'].sum()
dataplot.plot.bar(figsize = (20,10))
```

## The states producing loss are Arizona, Colarado, Florida, illinois, North Carolina, Ohio, Pennsylvania, Tennesee, Texas 

### Texas, Ohio, Illinois and Pennsylvania are incurring huge loss even when the sale is good

### New York and California are making the highest profit

```python
dataplot = df.groupby(['State']).Discount.sum()
plt.xlabel('States')
plt.ylabel('Discount')
dataplot.plot.bar(figsize = (20,10))
```

### Texas is giving very high discounts and thus incurring huge loss. The same is the case with Illinois, Ohio and Pennsylvania.
